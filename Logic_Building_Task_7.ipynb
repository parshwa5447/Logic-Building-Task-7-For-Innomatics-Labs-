# Problem 1: Smart Light Controller
### Problem Statement
Create a class to control a smart light that can be turned ON or OFF and display its status.
### üåç Real-Time Use
Smart home automation systems
### üì• Sample Input
Light Name: Bedroom Light \
Action: ON
### üì§ Expected Output
Bedroom Light is ON


```python
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF" #Default
    def turn_on(self):
        self.status = "ON"
    def turn_off(self):
        self.status = "OFF"
    def display_status(self):
        print(f"{self.light_name} is {self.status}")
light_name = input("Enter Light Name: ")
action = input("Enter Action (ON/OFF): ")
light = SmartLight(light_name)
if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()
else:
    print("Invalid Action!")
light.display_status()
```

# Problem 2: Employee ID Card System
### Problem Statement
Create a class that stores employee name, ID, and department and displays ID card details.
### Real-Time Use
Corporate HR systems
### üì• Sample Input
Employee Name: Rahul \
Employee ID: EMP102 \
Department: IT
### üì§ Expected Output
Employee ID Card \
Name: Rahul \
ID: EMP102 \
Department: IT

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    def display_id_card(self):
        print("\nEmployee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)
name = input("Enter Employee Name: ")
emp_id = input("Enter Employee ID: ")
department = input("Enter Department: ")
employee = Employee(name, emp_id, department)
employee.display_id_card()
```

# Problem 3: Mobile Contact Record
### Problem Statement
Create a class to store contact name and phone number and display contact information.
### Real-Time Use
Mobile contact applications
### üì• Sample Input
Contact Name: Anita \
Phone Number: 9876543210
### üì§ Expected Output
Contact Saved \
Name: Anita \
Phone: 9876543210


```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    def display_contact(self):
        print("\nContact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)
name = input("Enter Contact Name: ")
phone = input("Enter Phone Number: ")
contact = Contact(name, phone)
contact.display_contact()
```

# Problem 4: Product Price Tag Generator
### Problem Statement
Create a class that stores product name and price and prints a formatted price tag.
### Real-Time Use
Retail billing systems
### üì• Sample Input
Product Name: Headphones \
Price: 2499
### üì§ 0Expected Output
Product: Headphones \
Price: ‚Çπ2499


```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def display_price_tag(self):
        print("\nProduct:", self.name)
        print("Price: Rs." + str(self.price))
name = input("Enter Product Name: ")
price = float(input("Enter Price: "))
product = Product(name, price)
product.display_price_tag()
```

# Problem 5: Movie Rating Display System
### Problem Statement
Create a class that stores movie name and rating and displays movie details.
### Real-Time Use
Streaming platforms
### üì• Sample Input
Movie Name: Inception \
Rating: 4.8


```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    def display_movie_details(self):
        print("\nMovie Details")
        print("Movie Name:", self.name)
        print("Rating:", self.rating)
name = input("Enter Movie Name: ")
rating = float(input("Enter Rating: "))
movie = Movie(name, rating)
movie.display_movie_details()
```
